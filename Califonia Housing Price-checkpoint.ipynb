```python
#Define custom ratio function
def column_ratio(X):
    return X[:, [0]] / X[:, [1]]  # ratio of first column / second column

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]  # output feature name

def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler()
    )
```

```python
# Define log pipeline - log(1+x)
log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log1p, feature_names_out="one-to-one"),
    StandardScaler()
)
```

```python
# Default numeric pipeline
default_num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
import numpy as np
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

# ----------------------------
# Custom transformers
# ----------------------------

# 1 Ratio of MedInc / Population
def column_ratio(X):
    return (X[:, 0] / X[:, 1]).reshape(-1, 1)  # ensure 2D output

def ratio_name(function_transformer, feature_names_in):
    return ["income_per_person_ratio"]

ratio_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(column_ratio, feature_names_out=ratio_name)
)

# 2 Safe log transform
def safe_log1p(X):
    X_safe = np.maximum(X, 0)  # replace negatives with 0
    return np.log1p(X_safe)

log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(safe_log1p, feature_names_out="one-to-one")
)

# ----------------------------
# ColumnTransformer
# ----------------------------
full_preprocessing = ColumnTransformer([
    ("ratio", ratio_pipeline, ["MedInc", "Population"]),
    ("log", log_pipeline, ["MedInc", "AveOccup", "Population"]),
    ("passthrough", "passthrough", ["AveOccup"])  # optional, if you want raw AveOccup
],
remainder="drop")  # drop any other columns not listed

# ----------------------------
# Random Forest pipeline
# ----------------------------
rf_model = make_pipeline(
    full_preprocessing,
    RandomForestRegressor(
        n_estimators=50,
        max_depth=10,
        random_state=42,
        verbose=1
    )
)

# ----------------------------
# Fit and predict
# ----------------------------
rf_model.fit(X_train, y_train)                # fits the entire pipeline
rf_predictions = rf_model.predict(X_test)    # automatically transforms X_test

# Optional: print rounded predictions
print(rf_predictions[:5].round(-2))
```

```python
#################CODE FROM HERE#############
from sklearn.model_selection import cross_val_score
import numpy as np

# Example: 5-fold cross-validation using Random Forest
scores = cross_val_score(
    rf_model,       # your pipeline
    X_train,        # training features
    y_train,        # training target
    cv= 3,           # 3-fold CV (PC cant handle the load)
    scoring='neg_mean_squared_error'  # for regression
)

# Convert to RMSE
rmse_scores = np.sqrt(-scores)
print("RMSE per fold:", rmse_scores)
print("Mean RMSE:", rmse_scores.mean())
```
